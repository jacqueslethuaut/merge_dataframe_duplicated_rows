```python
import pandas as pd
import deduplicate
```

```python
df = pd.read_csv('Data/LPARS_sample.csv', sep=';')
df.head()
```

```python
df.info()
```

```python
c_date = 'Last Update Success'
c_key = 'Name'
date_format = '%Y-%m-%d %H:%M:%S.%f%z'
```

```python
df[c_date] = pd.to_datetime(df[c_date], format=date_format)
df.info()
```

```python
df = deduplicate.remove_duplicates(df, col_key=c_key, col_date=c_date)
```
